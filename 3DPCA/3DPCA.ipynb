```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
s_reset = pd.read_csv('s_reset.csv')
```

```python
s_reset = s_reset.drop(columns='Unnamed: 0')
```

```python
s_reset
```

```python
# 假设 df 是您的DataFrame
# df = pd.read_csv('your_data.csv')  # 如果您的数据在CSV文件中

# 标准化数据
scaler = StandardScaler()
df_scaled = scaler.fit_transform(s_reset)

# 应用PCA
pca = PCA(n_components=3)  # 降维到三维
pca_result = pca.fit_transform(df_scaled)

# 将结果转换为DataFrame
df_pca = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2', 'PC3'])

# 查看降维后的数据
print(df_pca)
```

```python
import plotly.graph_objs as go
import plotly.offline as pyo

x=list(df_pca['PC1'])
y=list(df_pca['PC2'])
z=list(df_pca['PC3'])

# 创建散点图
trace = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=12,
        color=z,  # 设置颜色为z轴的值
        colorscale='Viridis',  # 颜色映射
        opacity=0.8
    )
)

# 定义布局
layout = go.Layout(
    title='PCA result',
    scene=dict(
        xaxis=dict(title='PC1'),
        yaxis=dict(title='PC2'),
        zaxis=dict(title='PC3'),
    )
)

# 组合图表
fig = go.Figure(data=[trace], layout=layout)

# 显示图表
pyo.plot(fig, filename='PCA_result.html')
```

```python
import pandas as pd

# 假设 df1 和 df2 是您的两个DataFrame
# df1 = pd.read_csv('data1.csv')  # 包含24个变量
# df2 = pd.read_csv('data2.csv')  # 包含3个变量

# 确保两个DataFrame具有相同数量的行
df1 = s_reset
df2 = df_pca

# 合并DataFrame
df_combined = pd.concat([df1, df2], axis=1)

# 计算相关系数
corr_matrix = df_combined.corr()

# 提取与df2中的变量相关的系数
correlations = [corr_matrix[var].drop(var) for var in df2.columns]

# 如果需要，可以将结果转换为列表
correlations_lists = [corr.values.tolist() for corr in correlations]

# correlations_lists 中的每个列表将包含与df2中一个变量相关的24个变量的相关系数
```

```python
import plotly.graph_objs as go
import plotly.offline as pyo

x=correlations_lists[0]
y=correlations_lists[1]
z=correlations_lists[2]

names = [title for title in s_reset]

# 创建散点图
trace = go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=12,
        color=z,  # 设置颜色为z轴的值
        colorscale='Viridis',  # 颜色映射
        opacity=0.8
    ),
    text=names,  # 设置文本标签
    textposition='top center'  # 文本标签的位置
)

# 定义布局
layout = go.Layout(
    title='var_PCA corr',
    scene=dict(
        xaxis=dict(title='PC1'),
        yaxis=dict(title='PC2'),
        zaxis=dict(title='PC3'),
    )
)

# 组合图表
fig = go.Figure(data=[trace], layout=layout)

# 显示图表
pyo.plot(fig, filename='var_correlation.html')
```
