```python
import pandas as pd
df_para = pd.read_csv("para_20.csv")
```

```python
df_para = df_para.drop(columns='Unnamed: 0')
```

```python
# Assuming df is your DataFrame
# df = pd.DataFrame(...)

# Function to normalize a single column using robust scaling
def normalize_column(column):
    median = column.median()
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1

    # Subtract the median and divide by the IQR, then clip to [0, 1]
    return ((column - median) / IQR).clip(lower=0, upper=1)

# Apply the normalization to each column
df_normalized = df_para.apply(normalize_column)
```

```python
df_normalized = df_normalized.fillna(0)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 假设 df 是您的DataFrame
# df = pd.read_csv('your_data.csv')  # 如果您的数据在CSV文件中

# 标准化数据
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_normalized.iloc[:,:7])

# 应用PCA
pca = PCA(n_components=3)  # 降维到三维
pca_result = pca.fit_transform(df_scaled)

# 将结果转换为DataFrame
df_pca_consumer = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2', 'PC3'])

# 查看降维后的数据
print(df_pca_consumer)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# 假设 df 是您的DataFrame
# df = pd.read_csv('your_data.csv')  # 如果您的数据在CSV文件中

# 标准化数据
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df_normalized.iloc[:,7:])

# 应用PCA
pca = PCA(n_components=3)  # 降维到三维
pca_result = pca.fit_transform(df_scaled)

# 将结果转换为DataFrame
df_pca_seller = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2', 'PC3'])

# 查看降维后的数据
print(df_pca_seller)
```

```python
import plotly.graph_objs as go
import plotly.offline as pyo
import numpy as np

positives = np.array(list(df_para['pos_no']))

# 创建散点图
trace = go.Scatter3d(
    x=df_pca_consumer['PC1'],
    y=df_pca_consumer['PC2'],
    z=df_pca_consumer['PC3'],
    mode='markers',
    marker=dict(
        size=2,
        color=positives,  # 设置颜色为z轴的值
        colorscale='Blues',  # 颜色映射
        opacity=0.8
    )
)

# 定义布局
layout = go.Layout(
    title='PCA result',
    scene=dict(
        xaxis=dict(title='PC1'),
        yaxis=dict(title='PC2'),
        zaxis=dict(title='PC3'),
    ),
    paper_bgcolor='rgba(0,0,0,0)',
    plot_bgcolor='rgba(0,0,0,0)'
)

# 组合图表
fig = go.Figure(data=[trace], layout=layout)

# 显示图表
fig.write_html('comsumer.html')
```
